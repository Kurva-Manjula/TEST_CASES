# Which month have seen the most number of cancellation due to bad weather?

```python
weather_cancellation=df[df['CancellationCode']=='B']
```

```python
weather_cancellation_count=weather_cancellation.groupby('Month').size()
```

```python
weather_cancellation_count
```

```python
most_cancellation_month=weather_cancellation_count.idxmax()
```
